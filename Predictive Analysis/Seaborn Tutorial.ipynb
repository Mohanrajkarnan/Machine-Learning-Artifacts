```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
female_tips = tips.loc[(tips['sex']=='Female'),:]
male_tips = tips.loc[(tips['sex']!='Female'),:]
tips['sex'].value_counts()
```

## sns.relplot() - Relation between two variables

```python
sns.relplot(x='total_bill',y='tip',data=male_tips,style='time',hue='time')
pd.crosstab(tips['sex'],tips['time'])
```

```python
#consolidate
sns.relplot(x='total_bill',y='tip',data=tips,hue= 'sex',style='time',size='sex')
```
